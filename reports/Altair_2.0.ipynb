```python
import altair as alt
import pandas as pd


alt.renderers.enable('default')
```

```python
df = pd.read_excel("Sustainability_cleaned.xlsx",parse_dates=['Year'])
df.head()
column_names = df.columns
# parse_dates=['Year']
print(column_names)
```

```python
df = df[df["Country"]=="Finland"]
df.head()
```

```python
df = df.rename(columns={'Primary_school_enrol': 'Primary','Secondary_school_enrol':'Secondary'})

alt.Chart(df).mark_line(point=True).transform_fold(
    fold=['Primary', 'Secondary'],
    as_=['variable', 'value']
).encode(
    alt.X('Year', title = None),
    alt.Y('max(value):Q', title="Enrolment (%)", scale=alt.Scale(zero=False)),
    color=alt.Color('variable:N', legend=alt.Legend(title='School Enrolment')),
    tooltip=[alt.Tooltip('Year:T', format='%Y'), alt.Tooltip('max(value):Q', title='Literacy Level(%)')]
).properties(title="Literacy Level")
```

```python
df = df.rename(columns={'Unemployment_rate_male': 'Male','Unemployment_rate_women':'Female'})

alt.Chart(df).mark_line(point=True).transform_fold(
    fold=['Male', 'Female'],
    as_=['variable', 'value']
).encode(
    alt.X('Year', title =None),
    alt.Y('max(value):Q', title="Unemployment Rate(%)", scale=alt.Scale(zero=False)),
    color=alt.Color('variable:N', legend=alt.Legend(title='Gender')),
    tooltip=[alt.Tooltip('Year:T', format='%Y'), alt.Tooltip('max(value):Q', title='Unemployment Rate(%)')]   
).properties(title="Gender Gap in Unemployment")
```

```python
melted = pd.melt(df, id_vars=['Year'], value_vars=['Imports', 'Exports'], var_name='imp_exp', value_name='GDP_per_capita')
melted['Year'] = pd.DatetimeIndex(melted['Year']).year.astype(str)
# Create an Altair chart with the melted dataset, faceted by year
chart = alt.Chart(melted).mark_bar().encode(
    alt.X('imp_exp:N', axis=None), # remove x-axis label
    alt.Y('GDP_per_capita:Q', title="Trade(%) of GDP per capita"),
    alt.Color('imp_exp:N', legend=alt.Legend(title="Trade type", orient='bottom')),
    tooltip=[alt.Tooltip('GDP_per_capita:Q', title='GDP per capita'),alt.Tooltip('imp_exp:N', title='Trade type')]
).facet(
    column=alt.Column('Year:N', title=None)
).properties(
    title=alt.TitleParams(
        "Import and Export as % of GDP per Capita",
        anchor='middle',
        dy=-20
    )
)

# Show the chart
chart

```

```python
import altair as alt
import pandas as pd

# Load the data from the CSV file
data = pd.read_excel('Sustainability.xlsx')

# Load the world map information
world_map = alt.topo_feature('https://vega.github.io/vega-datasets/data/world-110m.json', 'countries')

# Create the chart
chart = alt.Chart(world_map).mark_geoshape(stroke='white', strokeWidth=0.5).encode(
    color='WSI:Q',
    tooltip=['Country:N', 'WSI:Q']
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(data, 'id', ['WSI', 'Country'])
).properties(
    width=800,
    height=500
)

chart
```

```python
import altair as alt
import pandas as pd

# Load data from CSV
df = pd.read_excel("Sustainability.xlsx")

# Define the chart
chart = alt.Chart(
    alt.topo_feature("https://raw.githubusercontent.com/deldersveld/topojson/master/world-110m.json", "countries")
).mark_geoshape(
    stroke="white",
    strokeWidth=0.5
).encode(
    color=alt.Color("WSI:Q", scale=alt.Scale(scheme="greens")),
    tooltip=["Country:N", "WSI:Q"]
).transform_lookup(
    lookup="id",
    from_=alt.LookupData(df, "Country", ["WSI"])
).properties(
    width=800,
    height=500
).project(
    type="naturalEarth1"
)

chart
```
